23) Faça um Scatter plot entre SepalLengthCm e PetalLengthCm

```python
import matplotlib.pyplot as plt

plt.scatter(df['SepalLengthCm'], df['PetalLengthCm'])
plt.xlabel('Sepal Length (Cm)')
plt.ylabel('Petal Length (Cm)')
plt.title('Scatter Plot entre as colunas Sepal Length (Cm) e Petal Length (Cm)')
plt.show()
```

24) Faça um Scatter plot entre SepalWidthCm e PetalWidthCm

```python
plt.scatter(df['SepalWidthCm'], df['PetalWidthCm'])
plt.xlabel('Sepal Width (Cm)')
plt.ylabel('Petal Width (Cm)')
plt.title('Scatter Plot entre as colunas Sepal Width (Cm) e Petal Width (Cm)')
plt.show()
```

25) Faça a tabela de Correlação entre as colunas

```python
import seaborn as sns

correlacao = df.corr()
sns.heatmap(correlacao, annot=True, cmap='coolwarm')
plt.title('Tabela de Correlação')
plt.show()
```
